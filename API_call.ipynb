```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import requests
import time
import json

# Import the OpenWeatherMap API key
from config import open_key

weather = pd.read_csv('top_cities.csv')
weather.head()
```

```python
#extracting columns needed
weather = weather[["city_ID", "city", "lat", "lng"]]

# #getting a sample size for testing
weather_df = weather.loc[weather["city_ID"]]



#creating extra columns for data
weather_df["Min Temp"] = ''
weather_df["Max Temp"] = ''
weather_df["Humidity"] = ''
weather_df["Cloudiness"] = ''

weather_df
```

```python
base_url = 'https://api.openweathermap.org/data/2.5/weather?'
#changing units to fahrenheit
units = 'Imperial'


for index, row in weather_df.iterrows():
    # get lat, lng from practice dataframe
    lat = row["lat"]
    lng = row["lng"]

    #creating url call
    temp_response = requests.get(f'{base_url}lat={lat}&lon={lng}&appid={open_key}&units={units}')
    temp_json = temp_response.json()

    #inserting data into the 
    try:
        weather_df.loc[index, "Min Temp"] = temp_json["main"]["temp_min"]
        weather_df.loc[index, "Max Temp"] = temp_json["main"]["temp_max"]
        weather_df.loc[index, "Humidity"] = temp_json["main"]["humidity"]
        weather_df.loc[index, "Cloudiness"] = temp_json["clouds"]["all"]
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
```

```python
weather_df

weather_df.to_csv('K_Resources/weather_data.csv', encoding='utf-8', index=False)
```
