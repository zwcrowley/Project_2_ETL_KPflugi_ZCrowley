```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import requests
import time
import json

# Import the OpenWeatherMap API key
from config import open_key

weather_df = pd.read_csv('K_Resources/top_cities.csv')
weather_df.head()
```

```python
#extracting columns needed
sample_data = weather_df[['city', 'lat', 'lng']].copy()

#getting a sample size for testing
practice_df = sample_data.sample(n = 5)

#creating extra columns for data
practice_df['Min Temp'] = ''
practice_df['Max Temp'] = ''
practice_df['Humidity'] = ''
practice_df['Clouds'] = ''

practice_df
```

```python
base_url = 'https://api.openweathermap.org/data/2.5/weather?'
#changing units to fahrenheit
units = 'Imperial'


for index, row in practice_df.iterrows():
    # get lat, lng from practice dataframe
    lat = row["lat"]
    lng = row["lng"]

    #creating url call
    temp_response = requests.get(f'{base_url}lat={lat}&lon={lng}&appid={open_key}&units={units}')
    temp_json = temp_response.json()

    #inserting data into the 
    try:
        practice_df.loc[index, "Min Temp"] = temp_json["main"]["temp_min"]
        practice_df.loc[index, "Max Temp"] = temp_json["main"]["temp_max"]
        practice_df.loc[index, "Humidity"] = temp_json["main"]["humidity"]
        practice_df.loc[index, "Clouds"] = temp_json["main"]["temp_min"]
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
```
